```python
import os
# Change directory to the desired path
os.chdir("C:/Users/guill/OneDrive/Documents/GitHub/Project-II/Serial/results/")

# Read and parse the data from the file
with open("g_r.dat") as f:
    data = f.readlines()

# Initialize lists to store x and y values
x = np.linspace(0.5*0.1, 0.5*0.01 + 0.01*(len(data)-1), len(data))
y = []

# Process each line of the data
for line in data:
    # Split the line into parts based on whitespace
    parts = line.split()
    # If there are enough parts (at least two) and the first part can be converted to a float
    if len(parts) >= 2 and parts[0].strip() and parts[1].strip():
        # Convert the first and second parts to floats and append them to x and y lists

        y.append(float(parts[1]))

y = y[::10]
x = x[::10]

y_max = max(y)
y = [i/y_max for i in y]

# Plot the data
plt.plot(x, y)
plt.xlabel('r')
plt.ylabel('g(r)')
plt.title('Radial distribution function')
plt.show()
```
