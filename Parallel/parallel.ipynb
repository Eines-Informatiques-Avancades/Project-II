```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
with open('time_vs_nproc.txt') as f:
    lines = f.readlines()
    
nproc = []
time = []

for line in lines:
# skip first line
    if line.startswith('nproc'):
        continue
    nproc.append(int(line.split()[0]))
    time.append(float(line.split()[1]))

df = pd.DataFrame({'nproc': nproc, 'time': time})
df = df.sort_values(by='nproc')

plt.plot(df['nproc'], df['time'], 'o-')
plt.xlabel('Number of processors')
plt.ylabel('Time (s)')
plt.title('Time vs Number of processors')
plt.grid()
plt.show()
```
